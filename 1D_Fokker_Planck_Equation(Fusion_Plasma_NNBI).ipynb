<h4>1.Fokker-Planck operator for small angle binary collisions of test particles in a magnetized Maxwellian plasma can be expressed as follows(Refs, [1][3]): </h4>
<h3>
$\left.\frac{\partial f}{\partial t}\right|_c=\frac{1}{\tau_s v^2} \frac{\partial}{\partial v}\left(v^3+v_c^3\right) f+\frac{1}{\tau_s v^2} \frac{\partial}{\partial v}\left(\frac{v^2 T_e}{m_b}+\frac{v_c^3 T_i}{m_b v}\right) \frac{\partial f}{\partial v} + \frac{v_{ii}}{2}\frac{\partial}{\partial \zeta}(1-\zeta^2)\frac{\partial}{\partial}f \hspace{4mm}(1)$
</h3>

<h4 >2. Simplified form of 1D-FP, ignoring the influence of the pitch angle : </h4>
<h3>$\left.\frac{\partial f}{\partial t}\right|_c=\frac{1}{\tau_s v^2} \frac{\partial}{\partial v}\left(v^3+v_c^3\right) f+\frac{1}{\tau_s v^2} \frac{\partial}{\partial v}\left(\frac{v^2 T_e}{m_b}+\frac{v_c^3 T_i}{m_b v}\right) \frac{\partial f}{\partial v} \hspace{4mm}(2)$
</h3>

Without CX loss and the Influence of an electric field, the equation can be expressed as:
<h3>$\frac{\partial f}{\partial t} = S_b + \left.\frac{\partial f}{\partial t}\right|_c \hspace{4mm}(3)$
</h3>
$S_b$ is the source term. The final form of equation to be discussed here can be expressed as:
<h3>$\frac{\partial f}{\partial t}=\frac{1}{\tau_s v^2} \frac{\partial}{\partial v}\left(v^3+v_c^3\right) f+\frac{1}{\tau_s v^2}  \frac{\partial}{\partial v}\left[\frac{T_e}{m_b v}\left( v^3 +\frac{T_i}{T_e}v_c^3\right)\frac{\partial f}{\partial v}\right]+S_b \hspace{4mm}(4)$
</h3>
The transformation of the second term on right side is performed with assuming an axisymmetric reactor and magnetic field($v \cdot \nabla f = 0$ and $v\times B \cdot \frac{\partial f}{\partial v} = 0$)


Denoting the beam energy at the time of injection as $E_{bo}[eV]$, according to $E_{bo} = \frac{1}{2}m_bv_{bo}^2$, $v$ can be normalized to $v' = v/v_{bo}$, We have

<h3>$\frac{1}{\tau_s v^2}\frac{\partial}{\partial v}(v^3+v_c^3)f = \frac{1}{\tau_s v'^{2}}\frac{\partial}{\partial v'}(v'^3+v_c'^3)f$</h3>

<h3>$\frac{1}{\tau_s v^2}  \frac{\partial}{\partial v}\left[\frac{T_e}{m_b v}\left( v'^3 +\frac{T_i}{T_e}v_c^3\right)\frac{\partial f}{\partial v}\right] = \frac{1}{\tau_s v'^2}  \frac{\partial}{\partial v'}\left[\frac{T_e}{m_b v_{bo}^2 v'}\left( v'^3 +\frac{T_i}{T_e}v_c'^3\right)\frac{\partial f}{\partial v'}\right] = \frac{1}{\tau_s v'^2}  \frac{\partial}{\partial v'}\left[\frac{T_e}{2E_{bo} v'}\left( v'^3 +\frac{T_i}{T_e}v_c'^3\right)\frac{\partial f}{\partial v'}\right]$</h3>

Thus, the function to discuss here can be normalized as:
<h3>
$\frac{\partial f}{\partial t} = \frac{1}{\tau_s v'^{2}}\frac{\partial}{\partial v'}(v'^3+v_c'^3)f + \frac{1}{\tau_s v'^2}  \frac{\partial}{\partial v'}\left[\frac{T_e}{2E_{bo} v'}\left( v'^3 +\frac{T_i}{T_e}v_c'^3\right)\frac{\partial f}{\partial v'}\right] + S_b \hspace{4mm}(5)$
</h3>

Where $v_c^{'3} = v_c^3/v_{bo}^3$

Since $\tau_s$ is very small($10^{-4}$ order in $n_i=n_e=1\times10^20$ and $T_i = T_e = 10\mathrm{keV}$), the advection and diffusion velocities are large. It is necessary to set a small time step. Normalizing $t$ by $\tau_s$ is a selection to avoid information loss errors and also simplify the selection of maximum time.

By applying $t' = \frac{t}{\tau_s}$, Eq.(5) can be normalized as above:

<h3>
$\frac{\partial f}{\partial t'} = \frac{1}{ v'^{2}}\frac{\partial}{\partial v'}(v'^3+v_c'^3)f + \frac{1}{v'^2}  \frac{\partial}{\partial v'}\left[\frac{T_e}{2E_{bo} v'}\left( v'^3 +\frac{T_i}{T_e}v_c'^3\right)\frac{\partial f}{\partial v'}\right] + \tau_s S_b \hspace{4mm}(6)$
</h3>

<h4 >3. Defining constants : </h4>

Notice that constants calculated by $cm$ order are all fixed into $m$ here.

Though irrelevant to this case, the generalized form of collision frequency corresponding to different ions(subscript i)is (Refs, [5])

$\tau_i = 12\pi^{3/2}\frac{\varepsilon_0^2m_i^{1/2}T_i^{3/2}}{n_iZ^4e^4\mathrm{ln}\Lambda_i}$

$\tau_s = \frac{6\pi\sqrt{2\pi}\epsilon_0^2m_em_b}{Z_b^2e^4n_eln{\Lambda_e}}(\frac{kT_e[K]}{m_e})^{\frac{3}{2}} \approx \frac{6.32\times10^{2}A_bT_e^{3/2}[eV]}{n_eZ_b^2ln\Lambda_e}$,

$v_{ii}=\frac{\langle Z \rangle \bar{A}_iv_c^3}{\tau_s[Z]A_bv^3}$,
    
$v_c^3 = 1.51 \times 10^{14}T_e^{3/2}[\mathrm{eV}] \frac{[Z]}{\bar{A}_i}$,


Notice that $v_c^3$ comes from a special form assuming only one kind of ion exists(Refs, [4]): $v_c = (0.75\pi^{1/2}m_e/m_i)^{1/3}v_e$ covering $\frac{[Z]}{\bar{A}_i}$.
    
$\langle Z\rangle = \frac{\Sigma_in_iZ_i^2ln\Lambda_i}{n_eln\Lambda_e}$    
    
$\frac{[Z]}{\bar{A}_i} = \frac{\Sigma_in_iZ_i^2ln\Lambda_i/A_i}{n_eln\Lambda_e}$


where $\bar{A}_i$ defined as the weighted average
 
$E_{bo} = 50 \mathrm{keV} $

This NBI energy $E_{bo} = 50 \mathrm{keV}$ does not satisfy any real case, though quite close to P-NB. To avoid relativity and the use of relativistic Fokker-Planck which we will discuss later.

P-NB NBI injection uses $E_{bo} = 85 \mathrm{keV}$. The post-incident beam energy is generated in a ratio of 7:2:1, each corresponding to an incident velocity of 1, 1/2, 1/3 out of $v_{\mathrm{bo}}$. This case leads to a different initial condition with three peaks. We will discuss it later.

For simplification, we assume to use only Deuterium in this case both for NBI and confined plasma in Tokamak.

Coulomb logarithm $log\Lambda_e$ is expressed as follows(Refs, [2])
For thermal electron-electron collisions:

$\begin{array}{lr}\ln \Lambda_c=23-\ln \left(n_e^{1 / 2} T_e^{-3 / 2}\right) & T_e<10 \mathrm{eV}, \\ \ln \Lambda_c=24-\ln \left(n_e^{1 / 2} T_e^{-1}\right) & T_e>10 \mathrm{eV} .\end{array}$

For thermal electron-ion collisions in this case, we have:

$\begin{array}{ll}\ln \Lambda_c=30-\ln \left(n_e^{1 / 2} T_i^{-3 / 2} Z_i^{3 / 2} \hat{m}_i\right) & T_e<T_i m_e / m_i, \\ \ln \Lambda_c=23-\ln \left(n_e^{1 / 2} Z_i T_e^{-3 / 2}\right) & T_i m_e / m_i<T_e<10 Z_i^2 \mathrm{eV}, \\ \ln \Lambda_c=24-\ln \left(n_e^{1 / 2} T_e^{-1}\right) & T_i m_e / m_i<10 Z_i^2 \mathrm{eV}<T_e .\end{array}$

$n_e$ and $n_i$ are measured of $\mathrm{cm}^{-3}$, $\hat{m}_i = m_i/m_p$ where $m_p$ is the proton mass

Because coulomb logarithms lie mostly between 10-20, it is considered to have a low impact on the solution. Thus, even though there are precise formulas for coulomb logarithms$\ln\Lambda$, we assume their values to be 17.


<h4>5.Boundary conditions:</h4>

The analysis of 1DFP for $f(v)$ tracks the incident particles, with the left boundary set at the velocity where incident particles lose energy and become part of the background plasma. Additionally, considering that velocities seldom diffuse beyond the initial values, it is assumed that a right boundary at $2v_{bo}$ with $f(v=v_{bo})=0$ is sufficient.

```python
def MLP(n_input,  n_output, n_neuron, n_layer, act_fn):
    input1 = keras.layers.Input(shape=(n_input,))
    input2 = keras.layers.Input(shape=(n_input,))


    merged = keras.layers.concatenate([input1, input2])
    hidden_layers=[]

    for _ in range(n_layer):
        hidden = keras.layers.Dense(
            units=n_neuron,
            activation=act_fn,
            kernel_initializer=tf.keras.initializers.GlorotNormal())
        hidden_layers.append(hidden)

    for hidden in hidden_layers:
            merged = hidden(merged)


    output = keras.layers.Dense(
            units=n_output,
            #activation="elu",
            #kernel_initializer=tf.keras.initializers.GlorotNormal()
            )(merged)

    model = tf.keras.models.Model(inputs=[input1, input2], outputs=output)
    return model
```

```python
class EarlyStopping:

    def __init__(self, patience=10, verbose=0):
        self.stopped = 0
        self.epoch = 0 
        self.pre_loss = float('inf') 
        self.patience = patience
        self.verbose = verbose # 早期終了メッセージの出力フラグをパラメーターで初期化

    def __call__(self, current_loss):

        if self.pre_loss <= current_loss: 
            self.epoch += 1 

            if self.epoch > self.patience:
                self.epoch = 0
                self.pre_loss = current_loss 
                return True

        else: 
            self.epoch = 0 
            self.pre_loss = current_loss 

        return False
```

To make autodiff calculations possible, further calculations are applied to target Eq.(6):
<h4>$ \frac{\partial f}{\partial t} = 3f + \left [\frac{1}{v^2}(v^3 +v_c^3) \right]\frac{\partial f}{\partial v}  + \frac{1}{ v^2} \left[ \frac{T_e v}{E_{bo}} - \frac{T_i v_c^3}{2E_{bo}v^2} \right] \frac{\partial f}{\partial v} + \frac{1}{ v^2} \left[ \frac{T_e v^2}{2E_{bo}} + \frac{T_i v_c^3}{2E_{bo}v}\right]\frac{\partial ^2 f}{\partial v^2} + \tau_s S_b -\tau_s \frac{f}{\tau_L}\hspace{4mm}(8)$</h4>

<h4>$\frac{\partial f}{\partial t} = 3f + \left [ v +\frac{v_c^3}{v^2} + \frac{T_e}{vE_{bo}} - \frac{T_i v_c^3}{2E_{bo}v^4} \right] \frac{\partial f}{\partial v} +  \left[ \frac{T_e }{2E_{bo}} + \frac{T_i v_c^3}{2E_{bo}v^3}\right]\frac{\partial ^2 f}{\partial v^2} + \tau_s S_b -\tau_s \frac{f}{\tau_L}\hspace{4mm}(9)$</h4>

$t$ is already normalized to $\frac{t}{\tau_s}$.

$v$ and $v_c$ normalized to $\frac{v}{v_{bo}}$ and $\frac{v_c}{v_{bo}}$

```python
class PhysicsInformedNNs():

    def __init__(self, n_input, n_output, n_neuron, n_layer, epochs,  act_fn='tanh'):
        self.batch = []
        self.loss_loc = []
        self.n_input = n_input
        self.n_output = n_output
        self.n_neuron = n_neuron
        self.n_layer = n_layer
        self.epochs = epochs
        self.act_fn = act_fn
        self.min_learning_rate = 1.0e-8

    def build(self, optimizer, loss_fn1, loss_fn2, early_stopping, factor):
        
        self._model = MLP(self.n_input, self.n_output, self.n_neuron, self.n_layer, self.act_fn)
        self._optimizer = optimizer
        self._loss_fn1 = loss_fn1
        self._loss_fn2 = loss_fn2
        self._early_stopping = early_stopping
        self._model_weights_save = [0, 0, self._model.get_weights()]
        self._loss_min = float('inf')
        self.factor = factor
        return self

    def train_step(self, t_boundary, v_boundary, t_initial, v_initial, t_pinn, v_pinn,v_pinn2,v_pinn3,v_pinn4, f_boundary, f_initial, tau_L_pinn, v_c3, T_i, T_e,  E_bo,loss_weights, loop):

        with tf.GradientTape() as tape_total:
            tape_total.watch(self._model.trainable_variables)

            # boundaries
            f_pred = self._model([t_boundary, v_boundary])
            loss1 = self._loss_fn1(f_pred, f_boundary)
            loss1 =  loss_weights[0]*tf.cast(loss1, dtype=tf.float32) # loss at boundaries
           
            #initials
            f_pred = self._model([t_initial, v_initial])
            loss2 = self._loss_fn1(f_pred, f_initial)
            loss2 = loss_weights[1]* tf.cast(loss2, dtype=tf.float32)
            
            # main domain
            with tf.GradientTape() as tapet:  
                tapet.watch(t_pinn)
                f_pred_pinn = self._model([t_pinn, v_pinn])
            f_t = tapet.gradient(f_pred_pinn, t_pinn)
            
            with tf.GradientTape() as tapev:
                tapev.watch(v_pinn)
                with tf.GradientTape() as tapevv:
                    tapevv.watch(v_pinn)
                    f_pred_pinn = self._model([t_pinn, v_pinn])
                f_v = tapevv.gradient(f_pred_pinn, v_pinn)
            f_vv =tapev.gradient(f_v, v_pinn)
            
            f_pred_pinn = self._model([t_pinn, v_pinn])
            f_pred_pinn = tf.cast(f_pred_pinn ,dtype=tf.float32)
            
            f_vv = tf.cast(f_vv, dtype=tf.float32)
            f_v = tf.cast(f_v, dtype=tf.float32)
            f_t = tf.cast(f_t, dtype=tf.float32)
            
                
            a1 = 3 *v_pinn3 * f_pred_pinn    # 長いので収まらない、のでlossの右辺を別々に書きます
            a3 = ((T_e*v_pinn3/(2*E_bo))+(T_i*v_c3/(2*E_bo)))* f_vv
            a2 = (v_pinn4 + v_c3*v_pinn + T_e*v_pinn2/(E_bo ) -(T_i * v_c3/(2*E_bo*v_pinn))) * f_v
            loss_physics = v_pinn3 * f_t - a1 - a2 - a3 + tau_L_pinn*f_pred_pinn*v_pinn3
            loss3 = loss_weights[2] * self._loss_fn1(loss_physics, tf.zeros_like(loss_physics)) # MSE of loss_physics; can be replaced with loss_physics^2

            # total loss
            loss = loss1 + loss2 + loss3
            
        # Choose minimize
        self._optimizer.minimize(loss, self._model.trainable_variables, tape=tape_total)
        self._loss_values.append(loss)

        # save weights and biases of the best model at this point
        if loss < self._loss_min:
            self._loss_min = loss
            self._model_weights_save = [loop, loss, self._model.get_weights()]
            
#        if loop % 20 == 0:
#            self.batch.append(loop)
#            self.loss_loc.append(loss.numpy())
            
        if loop % 100 == 0:
            print("train_step :", self.loop,"loss1(boundary_loss):", loss1.numpy(), "loss2(initial_loss) :", loss2.numpy(), "loss3(physics_loss) :",loss3.numpy(),"loss :", loss.numpy())

        return self

    def train(self, t_boundary, v_boundary, t_initial, v_initial, t_pinn, v_pinn,v_pinn2,v_pinn3,v_pinn4, f_boundary, f_initial, tau_L_pinn, v_c3, T_i, T_e, E_bo,loss_weights):
        
        self.loop = 0
        self._loss_values = []
        self.stopped = 0
        
        for i in range(self.epochs):
            self.loop += 1
            self.train_step(t_boundary, v_boundary, t_initial, v_initial, t_pinn, v_pinn,v_pinn2,v_pinn3,v_pinn4, f_boundary, f_initial, tau_L_pinn, v_c3, T_i, T_e, E_bo,loss_weights, self.loop)
            if self._early_stopping(self._loss_values[-1]):
                self.stopped +=1
                if self._optimizer.learning_rate * self.factor >= self.min_learning_rate:
                    self._optimizer.learning_rate.assign(self._optimizer.learning_rate * self.factor)
                    print("current_learning_rate:" , self._optimizer.learning_rate)
                if self.stopped >30:
                    self._model.set_weights(self._model_weights_save[2]) # load weights and biases of the best model
                    print('\nLoss of the best model at {0}: {1} '.format(*self._model_weights_save))
                    print("early stopping")
                    break
                else:
                    continue
```

<h3>Comparisons with the finite difference method</h3>

$\frac{\partial f}{\partial t} = \frac{1}{ v^{2}}\frac{\partial}{\partial v}(v^3+v_c^3)f + \frac{1}{v^2}  \frac{\partial}{\partial v}\left[\frac{T_e}{2E_{bo} v}\left( v^3 +\frac{T_i}{T_e}v_c^3\right)\frac{\partial f}{\partial v}\right] + \tau_s S_b \hspace{4mm}(5)$

Equation (5) is the target, buttransformations are required to apply the boundary conditions.

The boundary should be placed at twice of the ion temperature$T_i$, being designed to have no diffusive flux at that point. Therefore, for the left boundary, we initially disregard the applicability limits and achieve this by adding a loss term $S_{\mathrm{Loss}} = -f/\tau_L$ to the right side of the equation. Here, $\tau_L$ is fixed at 0.001, and it is assumed that particles approaching the left boundary with significant losses will undergo a substantial transport to the bulk plasma.(Refs, [3][4])

Thus, the targeted equation can be expressed as:
$ \frac{\partial f}{\partial t} = \frac{1}{ v^{2}}\frac{\partial}{\partial v}(v^3+v_c^3)f + \frac{1}{ v^2}  \frac{\partial}{\partial v}\left[\frac{T_e}{2E_{bo} v}\left( v^3 +\frac{T_i}{T_e}v_c^3\right)\frac{\partial f}{\partial v}\right] + \tau_s S_b - \frac{f}{\tau_L}\hspace{4mm}(6)$

In the context of Physics-Informed Neural Networks (PINNs), it might not be necessary to arbitrarily place the left boundary, as you can directly incorporate this requirement into the network during the training process.

For simplification, we set

$V = v^3 + v_c^3$ ,

$D = \frac{T_e}{2E_{bo}v}(v^3 + \frac{T_i}{T_e}v_c^3)$

to have a simpler expression of Eq.(6):

$\frac{\partial f}{\partial t} = \frac{1}{v^2}\frac{\partial}{\partial v}(Vf) +\frac{1}{ v^2}\frac{\partial}{\partial v}\left( D \frac{\partial f}{\partial v}\right) +\tau_s S_b -\frac{f}{\tau_L} \hspace{4mm}(8) $

<h4>Notice that different finite-difference methods to calculate source term are also presented here, but in this case we assume it to be zero, meaning no continous injection. </h4>

<h4>Upwind</h4>

This simplest method with a little lower precision but stable

$f_i^{n+1}-f_i^n = -\frac{\Delta t}{ v_i^2 \Delta v}\left[V_if_i^n-V_{i+1}f_{i+1}^n\right] + \frac{\Delta t}{ v_i^2 \Delta v^2}\left[\frac{D_{i+2}+D_{i+1}}{2}(f^n_{i+2}-f_{i+1}^n)-\frac{D_{i+1}+D_{i}}{2}(f_{i+1}^n-f_{i}^n)\right] -\frac{f_i^{n+1}+f_i^n}{2\tau_L} \hspace{4mm}(10)$

```python
v = np.linspace(left_boundary, right_boundary, nv)
v = np.append(v, v[-1]+deltav)
t = np.linspace(0.0, max_t, nt)

tau_indice = np.argmin(np.abs(v - 0.1))
tau_L = tau_s*1000

V = v**3 + v_c3
D = T_e*(v**3 + T_i*v_c3/T_e)/(2*E_bo*v)

condition = (v <= 1.0 - epsilon/2) | (v >= 1.0 + epsilon/2)

print("max_t:", max_t)
print("nt:", nt, "nv:", nv)
print("dv/dt:", deltav/deltat)
if deltat/(deltav)**2 <= 1/2 and deltav/deltat > max(V):
    print("test passed")
else:
    print("error, please reset deltat and deltav")
```

```python
start_time_Upwind = datetime.datetime.now()

f = np.zeros_like(v)
f = np.where(condition, 0.0, 1/epsilon)
delta_function = np.argmin(abs(v-1))
f[delta_function] = 1/epsilon

keep1 = []
def ket1(o):    
    return deltat/(v[o]**2*deltav)

def ket2(o):
    return deltat/(v[o]**2*deltav**2)

for n in range(nt):
    fn = f.copy()
    keep1.append(fn)
    for i in range(1, nv-1):
        f[i]= (-1*ket1(i) * (V[i] * fn[i] - V[i+1] * fn[i+1])  +
               ket2(i) * (((D[i+2] + D[i+1]) / 2) *  (fn[i+2] - fn[i+1]) - (( D[i+1] + D[i]) / 2) * (fn[i+1] - fn[i]))
               + fn[i]-tau_L*fn[i]/2)/(1+(tau_L/2)) 
        
end_time_Upwind = datetime.datetime.now()
elapsed_time_Upwind = abs(start_time_Upwind-end_time_Upwind)
```

The Upwind scheme has strict constraints in the choice of $\Delta t$ and $\Delta v$ to satisfy CFL(Courant-Friedrichs-Lewy)  because of the unstable advection velocity. As time progresses, distribution flatten, then diffusion decreases, leading to numerical instability problem near the left boundary. The CFL condition is expressed as $\frac{\Delta v}{\Delta t} > c$, where $c$represents the physical velocity. This means that the velocity of information transportation must be faster than the physical velocity. However, in this case for advection diffusion eqaution, diffusion velocity somehow counters that of the advection and since Fokker-Planck equation has both advection velocity and diffusion velocity of non-constants, it is hard to evaluate if the selection of $\Delta v$ and $\Delta t$ satisfies CFL condition. Here we assume that the diffusion term doesnt exist

$f_i^{n+1}-f_i^n = -\frac{\Delta t}{ v_i^2 \Delta v}\left[V_if_i^n-V_{i+1}f_{i+1}^n\right]\hspace{4mm}(11) $

Next, Von Neumann stability analysis is applied(Refs, [6]):

Expressing rounding error with an Fourier transform at $t=0$ as :

$\varepsilon_j^0 = \sum\limits_{m} V_{(k) }^0e^{ikx}, \hspace{4mm}$ $x = j\Delta v \hspace{4mm} (j=0, \dots , J)$

$k = m \pi/l, \hspace{4mm} l=J\Delta v \hspace{4mm} (m =0, \dots, J)$

the amplitude matrix then is:

$\mathbf{G}(\Delta t, k) = \frac{\mathbf{E}^{n+1}(k)}{\mathbf{E}^n(k)}$

$k = m \pi/l \hspace{4mm} (m =0, \dots, J)$ where $J$ is the maximum time grid

Then apply Fourier transform to Eq.(11):

$\tilde{f}^{n+1}(k) - \tilde{f}^n(k) =\frac{\Delta t}{ v_i^2 \Delta v}\left[ -V_i\tilde{f}^n(k)+V_{i+1}\tilde{f}^n(k)e^{ik\Delta v}  \right]$

Thus the amplitude matrix can be expressed as:

$\mathbf{G}_1(k) =\mathbf{I}+\frac{\Delta t}{v_i^2 \Delta v}(-V[i]+V[i+1]e^{ik\Delta v})$

\begin{align}
|\mathbf{G}|^2 &= [1+g(i)(-V[i]+V[i+1]e^{ik\Delta v})]^2\\
&= [1-g(i)V[i]+g(i)V[i+1]\mathrm{cos}(k\Delta v) +ig(i)V[i+1]\mathrm{sin}(k\Delta v)]^2\\
& = [1-g(i)V[i]+g(i)V[i+1]\mathrm{cos}(k\Delta v)]^2 -[g(i)V[i+1]\mathrm{sin}(k \Delta v)]^2\\
& = 1-2g(i)V[i]+g^2V[i]^2 + 2g(i)(1-g(i)V[i])V[i+1]\mathrm{cos}(k \Delta v) + g^2(i)V[i+1]^2(\mathrm{cos}^2(k\Delta v)-\mathrm{sin}^2(k \Delta v))
\end{align}

where $g(i) = \frac{\Delta t}{v_i^2 \Delta v}$

Thus, |G|^2<1 here becomes the Courant condition. Although the euqation is complex, it still means that the transformation velocity of information must be greater than that of advection velocity. It is hard to calculate and may be impossible to choose a proper composition of $\Delta t$ and $\Delta v$ if diffusion term exits here because diffusion velocity has both direction and also change with time and velocity. proofs above show that simple Upwind scheme is not just inaccurate but even hard to have computational stability.

<h4>Simple Crank-Nicolson</h4> (Ref -> Advection_diffusion_equation.ipynb)

By introducing subscripts $i$ for spatial positions(velocity$v$ here) and $n$ for time, we can discretize Equation (6) using the Crank-Nicolson method and obtain the following expression:

<h4> 
\begin{align}
\frac{f_i^{n+1}-f_i^n}{\Delta t} &= \frac{1}{v_i^2}\frac{1}{2}\left[ \frac{V_{i+1}^{n+1}f_{i+1}^{n+1} - V_{i-1}^{n+1}f_{i-1}^{n+1}}{2\Delta v} + \frac{V_{i+1}^nf_{i+1}^n - V_{i-1}^nf_{i-1}^n}{2\Delta v}\right]\\
&= \frac{1}{ v_i^2}\frac{1}{2}\left[ \frac{D_{i+1}^{n+1}+D_i^{n+1}}{2}\frac{f_{i+1}^{n+1} - f_i^{n+1}}{\Delta v}- \frac{D^{n+1}_i+D^{n+1}_{i-1}}{2}\frac{f_i^{n+1}-f^{n+1}_{i-1}}{2\Delta v} \right]\frac{1}{\Delta v}\\
&= \frac{1}{ v_i^2}\frac{1}{2}\left[ \frac{D_{i+1}^{n}+D_i^{n}}{2}\frac{f_{i+1}^{n} - f_i^{n}}{\Delta v}- \frac{D^{n}_i+D^{n}_{i-1}}{2}\frac{f_i^{n}-f^{n}_{i-1}}{2\Delta v} \right]\frac{1}{\Delta v}\\
&+ \tau_s S_{bi}^n - \frac{f_i^{n+1}+f_i^n}{2\tau_L}
\end{align}</h4>

Arraging the equation by subscripts yields:

\begin{align}
-\frac{\Delta t}{4\Delta v v_i^2}\left(V^{n+1}_{i+1} + \frac{D_{i+1}^{n+1}+D_i^{n+1}}{\Delta v}\right)f_{i+1}^{n+1} 
+ \left[ 1+\frac{\Delta t}{4\Delta v v_i^2}\left(\frac{D_{i+1}^{n+1}+2D_i^{n+1}+D_{i-1}^{n+1}}{\Delta v}\right)+\frac{\Delta t}{2\tau_L}\right]f_i^{n+1} 
-\frac{\Delta t}{4\Delta v v_i^2}\left(-V^{n+1}_{i-1} + \frac{D_{i}^{n+1}+D_{i-1}^{n+1}}{\Delta v}\right)f_{i-1}^{n+1} \\
= \frac{\Delta t}{4\Delta v v_i^2}\left(V^{n}_{i+1} + \frac{D_{i+1}^{n}+D_i^{n}}{\Delta v}\right)f_{i+1}^{n} 
+ \left[-\frac{\Delta t}{2\tau_L} + 1 - \frac{\Delta t}{4\Delta v v_i^2}\left(\frac{D_{i+1}^{n}+2D_i^n+D_{i-1}^n}{\Delta v}\right)\right]f_i^n 
+ \frac{\Delta t}{4\Delta v v_i^2}\left(-V_{i-1}^n +\frac{D_{i}^{n}+D_{i-1}^{n}}{\Delta v}\right)f_{i-1}^n + \tau_s S_{bi}^n\Delta t
\end{align}

When using simplified coefficients, the equation above can be expressed as:
$A^{n+1}f_{i+1}^{n+1} + B^{n+1}f_i^{n+1} +C^{n+1}f_{i-1}^{n+1} = -A^nf_{i+1}^n +(2-B^n)f_i^n -C^nf_{i-1}^n +\tau_s S_{bi}^n\Delta t$

where

$A^n = - \frac{\Delta t}{4\Delta v v_i^2}\left(V^{n}_{i+1} + \frac{D_{i+1}^{n}+D_i^{n}}{\Delta v}\right)$ 

$B^n = \left[ 1+\frac{\Delta t}{4\Delta v v_i^2}\left(\frac{D_{i+1}^{n}+2D_i^{n}+D_{i-1}^{n}}{\Delta v}\right)+\frac{\Delta t}{2\tau_L}\right]$

$C^n = -\frac{\Delta t}{4\Delta v v_i^2}\left(-V^{n+1}_{i-1} + \frac{D_{i}^{n}+D_{i-1}^{n}}{\Delta v}\right) $

This can be derived into a matrix equation, which is able to be solved with Thomas's algorithm

\begin{array}
$\begin{bmatrix}
  A^{n+1} & 0 & 0 & 0 & \dots & 0\\
  B^{n+1} & A^{n+1} & 0 & 0 & \dots & 0\\
  C^{n+1} & B^{n+1} & A^{n+1} & 0 & \dots & 0\\
  0 & C^{n+1} & B^{n+1} & A^{n+1} & \dots & 0\\
  \vdots & \vdots & \vdots &  \vdots & \ddots & \vdots\\
  0 & 0 & \dots & C^{n+1} & B^{n+1} & A^{n+1}\\
  0 & 0 & 0 & \dots & C^{n+1} & B^{n+1}
\end{bmatrix}
\begin{bmatrix}
f_1^{n+1}\\
f_2^{n+1}\\
f_3^{n+1}\\
f_4^{n+1}\\
\vdots\\
f_{i-2}^{n+1}\\
f_{i-1}^{n+1}
\end{bmatrix}
\begin{matrix}
=
\end{matrix}
\begin{bmatrix}
  -A^n & 0 & 0 & 0 & \dots & 0\\
  (2-B^{n}) & -A^n & 0 & 0 & \dots & 0\\
  -C^n & (2-B^{n}) & -A^n & 0 & \dots & 0\\
  0 & -C^n & (2-B^{n}) & -A^n & \dots & 0\\
  \vdots & \vdots & \vdots &  \vdots & \ddots & \vdots\\
  0 & 0 & \dots & -C^n & (2-B^{n}) & -A^n\\
  0 & 0 & 0 & \dots & -C^n & (2-B^{n})
\end{bmatrix}
\begin{bmatrix}
f_1^{n}\\
f_2^{n}\\
f_3^{n}\\
f_4^{n}\\
\vdots\\
f_{i-2}^{n}\\
f_{i-1}^{n}
\end{bmatrix}
+
\begin{matrix}
\begin{bmatrix}
\tau_s S_{b0}^n\Delta t\\
\tau_sS_{b1}^n\Delta t\\
\tau_sS_{b2}^n\Delta t\\
\vdots\\
\tau_sS_{b(i-2)}^n\Delta t\\
\tau_sS_{b(i-1)}^n\Delta t\\
\end{bmatrix}
\end{matrix}\hspace{4mm}
\end{array}

```python
L = np.zeros((nv+1, nv+1))
R = np.zeros((nv+1, nv+1))

def A(i):
    result = (-1*deltat/(4*deltav*v[i-1]**2))*(V[i]+(D[i]+D[i-1])/deltav)
    return result

def B(i):
    result = 1+ deltat*(D[i+1]+2*D[i]+D[i-1])/(4*deltav**2*v[i]**2)
    return result
def B1(i):
    result = 1+ deltat*(D[i+1]+2*D[i]+D[i-1])/(4*deltav**2*v[i]**2)+deltat*tau_L/2
    return result
def C(i):
    result = -1*deltat*(-1*V[i] + (D[i+1]+D[i])/deltav)/(4*deltav*v[i+1]**2)
    return result
```

```python
f = np.zeros_like(v)
f = np.where(condition, 0.0, 1/epsilon)
f_new = f.copy()
keep2 = []
keep2.append(f.copy())
start_time_Crank_Nicolson = datetime.datetime.now()

for n in range(nt):
    for i in range(2, tau_indice+1):
        L[i, i-1] = C(i-1)
        L[i, i] = B1(i)
        L[i, i+1] = A(i+1)
            
        R[i, i-1] = -C(i-1)
        R[i, i] = 2 - B1(i)
        R[i, i+1] = -A(i+1)
    for i in range(tau_indice+1, nv):
        L[i, i-1] = C(i-1)
        L[i, i] = B(i)
        L[i, i+1] = A(i+1)
            
        R[i, i-1] = -C(i-1)
        R[i, i] = 2 - B(i)
        R[i, i+1] = -A(i+1)
        
    L[1,1] = L[2,2]    
    L[nv, nv-1] = L[nv-1, nv-2]
    L[nv, nv] = L[nv-1, nv-1]
            
    R[nv, nv-1] = R[nv-1, nv-2]
    R[nv, nv] = R[nv-1, nv-1]
 
    f_new[1:nv] = np.linalg.solve(L[1:nv, 1:nv], np.dot(R[1:nv, 1:nv], f[1:nv]))
    keep2.append(f_new.copy())
    f = f_new
    
end_time_Crank_Nicolson = datetime.datetime.now()
elapsed_time_Crank_Nicolson = abs(end_time_Crank_Nicolson-start_time_Crank_Nicolson)
```

<h4>Crank-Nicolson with Upwind</h4>

The target equation has a dominant advection term. Thus, using the Crank-Nicolson method for the advection term incorporates downstream information, leading to numerical instability. Introducing the Upwind scheme resolves the numerical instability issue, but also leads to the accuracy to drop down to first order.

<h4> 
\begin{align}
 \frac{f_i^{n+1}-f_i^n}{\Delta t} &= \frac{1}{v_i^2}\frac{1}{2}\left[ \frac{V_{i+1}f_{i+1}^{n+1} - V_{i}f_{i}^{n+1}}{\Delta v} + \frac{V_{i+1}f_{i+1}^n - V_{i}f_{i}^n}{\Delta v}\right]\\
&= \frac{1}{ v_i^2}\frac{1}{2}\left[ \frac{D_{i+1}+D_i}{2}\frac{f_{i+1}^{n+1} - f_i^{n+1}}{\Delta v}- \frac{D_i+D_{i-1}}{2}\frac{f_i^{n+1}-f^{n+1}_{i-1}}{\Delta v}
+ \frac{D_{i+1}+D_i}{2}\frac{f_{i+1}^{n} - f_i^{n}}{\Delta v}- \frac{D_i+D_{i-1}}{2}\frac{f_i^{n}-f^{n}_{i-1}}{\Delta v} \right]\frac{1}{\Delta v}\\
&+ \tau_sS_{bi}^n - \frac{f_i^{n+1}+f_i^n}{2\tau_L}
\end{align}</h4>

Notice that $V$ and $D$ dont change with time.

<h4>Crank-Nicolson UTOPIA(Uniformly Third-Order Polynomial Interpolation Algorithm)</h4>

UTOPIA is based on the concept to use polynomial approximation to evaluate truncation error. This method aims to provide a well balanced error between advection term and diffusion term.

With Chakravarthy-Osher scheme and k = 1/3, put $g = Vf$. The advection term on $v \leq 0$ can be written as[7]:

$\left(\frac{\partial g}{\partial v}\right)_i = \frac{1}{\Delta v}\left[-2g_{i-1}-3g_i+6g_{i+1}-g_{i+2}\right]$

By introducing subscripts $i$ for spatial positions(velocity$v$ here) and $n$ for time, we can discretize Equation (6) using the Crank-Nicolson method on diffusion term and UTOPIA method on advection term. This leads to the following expression:
